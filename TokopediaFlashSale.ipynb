```python
# url yang ingin diakses
url = "https://www.tokopedia.com/manifest.json"

# mengirim permintaan GET
response_tokopedia_cookies = requests.get(url)

if response_tokopedia_cookies.status_code == 200:
  cookies = response_tokopedia_cookies.cookies
  for cookie in cookies:
    print(f"name: {cookie.name}, Value: {cookie.value}, Domain: {cookie.domain}, Path: {cookie.path}")
else:
  print(f"requset failed with status code : {response_tokopedia_cookies.status_code}")
```

```python
# url API tokopedia
url = "https://gql.tokopedia.com/graphql/ComponentInfoQuery"

# Headers untuk request
headers = {
  "Referer": "https://www.tokopedia.com/discovery/kejar-diskon?rpc_sortfilter_cid=1_4400721&source=homepage.left_carousel.0.308281",
  "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/130.0.0.0 Safari/537.36"
}

cookies = {
    "_abck": response_tokopedia_cookies.cookies['_abck'],
    "bm_sz": response_tokopedia_cookies.cookies['bm_sz']
}

body = [{
    "operationName":"ComponentInfoQuery",
    "variables":
     {
         "identifier":"kejar-diskon",
         "componentId":"5",
         "device":"desktop",
         "filters":"{\"rpc_sortfilter_cid\":\"4400721\",\"rpc_UserID\":\"0\",\"count_only\":\"true\",\"rpc_UserAddressId\":\"128055257\",\"rpc_UserCityId\":\"218\",\"rpc_UserDistrictId\":\"2956\",\"rpc_UserLat\":\"-7.218836699999998\",\"rpc_UserLong\":\"112.09424\",\"rpc_UserPostCode\":\"62194\",\"rpc_UserWarehouseId\":\"16699633\",\"source\":\"homepage.left_carousel.0.308281\"}"
         },
    "query":"query ComponentInfoQuery($identifier: String!, $componentId: String!, $device: String!, $filters: String) {\n  componentInfo(identifier: $identifier, component_id: $componentId, device: $device, filters: $filters) {\n    data\n    __typename\n  }\n}\n"
}]

# mengirim request post ke API
response_flashsale = requests.post(url, headers=headers, cookies=cookies, json=body)

# cek jika response_flashsale berhasil
if response_flashsale.status_code == 200:
  data = response_flashsale.json()
  print("Data Retrived: ")
  print(json.dumps(data, indent=4))
else:
  print(f"Request failed with status code: {response_flashsale.status_code}")
  print(response_flashsale.text)
```

```python
def get_soup_from_url(url):
  options = Options()
  options.add_argument("--headless")  # Jalankan browser tanpa tampilan GUI
  options.add_argument("--no-sandbox")
  options.add_argument("--disable-dev-shm-usage")
  options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.88 Safari/537.36")

  # Inisialisasi WebDriver dengan ChromeDriverManager
  driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

  try:
    url ="https://www.tokopedia.com/movioofficialstore/urine-bag-gel-kantong-kencing-kantong-pipis-darurat-car-toilet-pack-isi-4-093cd?extParam=src%3Dmultiloc%26whid%3D10538060"
    driver.get(url)

    time.sleep(5)

    page_source = driver.page_source
    soup = BeautifulSoup(page_source, "html.parser")

    print(soup.prettify())

  except Exception as e:
    print(f"Terjadi kesalahan: {e}")

  finally:
    driver.quit()
```

```python

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/130.0.0.0 Safari/537.36"
}

product_detail_list = []

def get_soup_from_url(url):
    rsp = requests.get(url, headers=headers)
    if rsp.status_code == 200:
        return BeautifulSoup(rsp.content, "html.parser")
    return None

for product in tqdm(data[0]['data']['componentInfo']['data']['component']['data']):
    soup = get_soup_from_url(product['url_desktop'])
    if soup:
        rating = soup.find("span", attrs={"data-testid": "lblPDPDetailProductRatingNumber"})
        description = soup.find("div", attrs={"data-testid": "lblPDPDescriptionProduk"})

        product_data = {
            "product_id": product["product_id"],
            "name": product["name"],
            "rating": rating.text if rating else None,  # Safely extract text if rating exists
            "custom_stock": product.get("custom_stock"),
            "preorder": product.get("preorder"),
            "price": product.get("price"),
            "discount_precentage": product.get("discount_precentage"),  # Use .get() to handle missing keys
            "discounted_price": product.get("discounted_price"),
            "shop_name": product.get("shop_name"),
            "shop_location": product.get("shop_location"),
            "description": description.text if description else None,  # Safely extract text if description exists
            "url_desktop": product.get("shop_url_desktop"),
            "campaign_id": product.get("campaign_id"),
            "campaign_sold_count": product.get("campaign_sold_count"),
            "is_official": product.get("is_official"),
            "is_topads": product.get("is_topads")
        }

        product_detail_list.append(product_data)
```

```python
df_flashsale = pd.DataFrame(product_detail_list)

df_flashsale
```
